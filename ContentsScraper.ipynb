```python
list_1 = list()
list_2 = list()
list_3 = list()
list_4 = list()

for k,v in articles.items():
    list_1.append(v.title)
    list_2.append(v.url)
    list_3.append(v.maintext)
    list_4.append(v.date_publish)

print(list_1)
print(list_2)
print(list_3)
```

```python
df = pd.DataFrame(list(zip(list_1, list_2, list_3, list_4)),
               columns =['Title', 'URL', 'Text', 'Date'])
df
```
